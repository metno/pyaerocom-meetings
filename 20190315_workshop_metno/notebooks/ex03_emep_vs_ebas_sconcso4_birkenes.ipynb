```python
dataset = pya.io.ReadGridded('EMEP_rv4_2599_Rep2013emistrend')
print(dataset)
```

```python
emep_data = dataset.read_var('sconcso4', start=2000, stop=2011, ts_type='daily')
emep_data
```

Get timeseries at Birkenes

```python
stations = emep_data.to_time_series(latitude=[ebas_birkenes.latitude], 
                                    longitude=[ebas_birkenes.longitude])
```

```python
emep_at_birkenes = stations[0]
emep_at_birkenes
```

```python
ts_ebas = ebas_birkenes.to_timeseries('sconcso4', freq='monthly')['2000':'2010']
```

```python
ax = emep_at_birkenes.plot_timeseries('sconcso4', freq='daily')
ax = emep_at_birkenes.plot_timeseries('sconcso4', freq='monthly', ax=ax, lw=4)
ax = ts_ebas.plot(ax=ax, color='lime', label='EBAS')
ax.legend();
```

```python
import pandas as pd
ts_emep = emep=emep_at_birkenes.to_timeseries('sconcso4', freq='monthly')
merged = pd.concat(dict(ebas=ts_ebas, emep=ts_emep), axis=1)
merged.head()
```

```python
merged.min(), merged.max()
```

```python
ax = merged.plot.scatter('ebas', 'emep', xlim=(0, 3), ylim=(0,3))
ax.plot([0,3], [0,3], '--', c='k')
ax.grid()
```
